```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "/content/dog_mochi.png"
img = plt.imread(image_path)
plt.imshow(img)
```

```python
conv = nn.Conv2d(in_channels=3,
                out_channels=7,
                kernel_size=5)
```

```python
x = torch.from_numpy(img)
x = x.permute(2, 0, 1)
x = x.reshape([1, 3, 241, 145])
y = conv(x)
y = y.detach().numpy()
print(y.shape)
```

```python
y = y[0]
print(y.shape)
```

```python
y_max = np.max(y)
print(f"Y Max: {y_max}")
y_min = np.min(y)
print(f"Y Max: {y_min}")
```

```python
img_after_conv = y - y_min / (y_max - y_min)
print(img_after_conv.shape)
print(img_after_conv)
```

```python
plt.figure(figsize=(14,4))
for i in range(7):
    plt.subplot(1, 7, i+1)
    plt.imshow(img_after_conv[i])
```
