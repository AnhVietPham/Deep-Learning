```python
import math
import tensorflow as tf
from d2l import tensorflow as d2l
```

```python
eta = 0.1
```

```python
def constant_lr():
  return 1
```

```python
lr = constant_lr
```

```python
def f(x1, x2):
  return x1**2 + 2*x2**2
```

```python
def f_grad(x1, x2):
  return 2*x1, 4*x2
```

```python
def sgd(x1, x2, s1, s2, f_grad):
  g1, g2 = f_grad(x1, x2)
  g1 += tf.random.normal([1], 0.0, 1)
  g2 += tf.random.normal([1], 0.0, 1)
  eta_t = eta * lr()
  return (x1 - eta_t * g1, x2 - eta_t * g2, 0, 0)
```

```python
d2l.show_trace_2d(f, d2l.train_2d(sgd, steps=500, f_grad=f_grad))
```

```python
def exponential_lr():
  global t
  t += 1
  return math.exp(-0.1 * t)
```

```python
t = 1
lr = exponential_lr
d2l.show_trace_2d(f, d2l.train_2d(sgd, steps=1000, f_grad=f_grad))
```

```python
def polynomial_lr():
  global t
  t += 1
  return (1 + 0.1 * t) ** (-0.5)
```

```python
t = 1
lr = polynomial_lr
d2l.show_trace_2d(f, d2l.train_2d(sgd, steps=50, f_grad=f_grad))
```
